
6. Visualiza la distribución de la variable "cpi" utilizando un histograma y comenta sobre la distribución de los índices de percepción de corrupción.

```python
# Crear el histograma usando matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['cpi'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución del Índice de Percepción de la Corrupción (CPI)')
plt.xlabel('Índice de Percepción de la Corrupción (CPI)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Alternativa usando seaborn para un histograma más estilizado
sns.histplot(df['cpi'], kde=True, bins=10, color='skyblue')
plt.title('Distribución del Índice de Percepción de la Corrupción (CPI)')
plt.xlabel('Índice de Percepción de la Corrupción (CPI)')
plt.ylabel('Frecuencia')
plt.show()
```


7. Encuentra el país con la mayor diferencia entre el "gdp" (Producto Interno Bruto) y el "co2-emissions" (emisiones de dióxido de carbono).

```python
# Calcular la diferencia entre 'gdp' y 'co2_emissions'
df['diferencia'] = df['gdp'] - df['co2-emissions']

# Encontrar el país con la mayor diferencia
pais_max_diferencia = df.loc[df['diferencia'].idxmax()]

# Mostrar el país con la mayor diferencia
print(f"El país con la mayor diferencia entre el PIB y las emisiones de CO2 es: {pais_max_diferencia['country']}")
print(f"Diferencia: {pais_max_diferencia['diferencia']}")
```
