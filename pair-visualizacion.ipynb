```python
# 1. ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

sns.histplot(x= "birth_rate", data = df, bins = 10, kde=True)
#Tenemos más paises con tasa de natalidad baja que alta. 
# Por lo que podemos extraer que a nivel mundial la natalidad está entre 10 y 20.
```

```python
sns.violinplot(x= "birth_rate", data = df)
```

```python
#2. ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? 
# Proporciona una breve descripción de los patrones que observes.
sns.scatterplot(x = "life_expectancy", y="infant_mortality", data = df);
#Podemos apreciar que la mortalidad infantil tiene un impacto importante 
#en la esperanza de vida de los paises ya que a más alto es este ratio,
#menor la esperanza de vida.
```

```python
sns.regplot(x = "life_expectancy", y="infant_mortality", data = df);
```

```python
df_conteo= df["continent"].value_counts().reset_index()
df_conteo
```

```python
# 3. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? 
# Proporciona el porcentaje de países en ese continente.

plt.pie("continent", labels="index", data=df_conteo, autopct=  '%1.2f%%')
```

```python
plt.pie("continent", labels="index", data=df_conteo, autopct=  '%1.2f%%')
```

```python
sns.countplot(x= "continent", data = df);
plt.xticks(rotation = 90)
#El continente que mas paises tiene es Africa
```
